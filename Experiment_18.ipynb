Exploring Statistical and Specialized Data Visualization Techniques

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D' , 'E'],
    'Values': [0,70,30,90,60],
    'Sales':np.random.randint(100,500,5),
    'Profit':np.random.randint(10,100,5)
})
print(df)
```

```python
# Area Plot
plt.figure()
plt.fill_between(df['Category'], df['Values'], color='blue', alpha=0.25)
plt.title('Area Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.figure(figsize=(7, 4))
plt.fill_between(df['Category'], df['Sales'], color='blue', alpha=0.5,label = 'Sales')
plt.fill_between(df['Category'], df['Profit'], color='orange', alpha=0.5,label = 'Profit')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Pie Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Donut Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Navya')
plt.show()
```

```python
# BoxPlot
plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Boxplot of Sales")
plt.show()
```

```python
# HeatMap
plt.figure()
corr = df[['Sales','Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df[['Values', 'Sales','Profit']].corr()
plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#Bubble Plot
import matplotlib.pyplot as plt
plt.scatter(df['Sales'],df['Profit'],
            s = df['Values']*10)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Bubble Plot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.scatterplot(x = 'Sales', y = 'Profit', data = df, size = 'Values',hue = 'Values',sizes =(50,300), palette ='viridis')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```
